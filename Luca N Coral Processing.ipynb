```python
import numpy as np
import pandas as pd
import sklearn.neighbors
import sklearn.cluster
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./deep_sea_corals.gz')

df = df.drop(['DataProvider','Station','Locality','SurveyID',
             'Repository','EventID', 'SampleID', 'DepthMethod'], axis=1)
```

```python
skip = 5
long = df['longitude'][1::skip].astype('f16') * np.pi/180
lat = df['latitude'][1::skip].astype('f16') * np.pi/180

plt.figure(figsize=(10,5))
plt.plot(long, lat, marker='.', linestyle='')
```

```python
points = np.transpose(np.stack([lat, long]))

clust = sklearn.cluster.DBSCAN(eps=700/6371, min_samples=20, 
                               metric='haversine').fit(points)

clust.labels_.max()
```

```python
clust.labels_.min()
```

```python
model = sklearn.neighbors.KNeighborsClassifier(n_neighbors = 5, weights = 'distance',
                                               metric='haversine')

model = model.fit(points, clust.labels_)

pad_labels = np.full(df.shape[0]-1, -2)
pad_labels[np.arange(0, df.shape[0], skip)] = clust.labels_

mask = pad_labels == -2

new_points = np.stack([df['latitude'][1:].astype('f16')[mask],df['longitude'][1:].astype('f16')[mask]], axis=-1) * np.pi/180

new_labels = model.predict(new_points)

pad_labels[mask] = new_labels
pad_labels = np.insert(pad_labels, 0, -1)
```

```python
points = np.stack([df['longitude'][1:].astype('f16'),df['latitude'][1:].astype('f16')], axis=-1) * np.pi/180
groups = dict()
for i, g in enumerate(pad_labels[1:]):
    if g in groups:
        groups[g].append(tuple(points[i]))
    else:
        groups[g] = [tuple(points[i])]

if False:
    for key, val in groups.items():
        plt.figure(figsize=(5,2.5))
        plt.title(f'{key}: {len(val)}')
        plt.xlim(-np.pi,np.pi)
        plt.ylim(-np.pi/2,np.pi/2)
        plt.plot(*zip(*val), marker='.', linestyle='')
        plt.show()
else:
    plt.figure(figsize=(10,5))
    plt.xlim(-np.pi,np.pi)
    plt.ylim(-np.pi/2,np.pi/2)
    for key, val in groups.items():
        if key != -1:
            plt.plot(*zip(*val), marker='.', linestyle='', label=key)
    plt.legend(bbox_to_anchor=(1.1, 1.))
    plt.show()
```

```python
df['Depth>0'] = df['DepthInMeters'].apply(lambda x:max(x,0))
df['Expert'] = ~df['IdentificationQualifier'].apply(str).str.contains('(?i)non-expert')

df.at[417782, 'ObservationDate'] = '1954-5-26'

df.at[417786, 'ObservationDate'] = '1958-11-29'

coral_classify = {
    'nan': 'Other',
    'stony coral (branching)': 'Stony',
    'gorgonian coral': 'Stony',
    'stony coral (cup coral)': 'Stony',
    'black coral': 'Soft',
    'sea pen': 'Soft',
    'demosponge': 'Sponge',
    'lace coral': 'Soft',
    'sponge (unspecified)': 'Sponge',
    'stony coral (unspecified)': 'Stony',
    'stoloniferan coral': 'Soft', 
    'soft coral': 'Soft', 
    'glass sponge': 'Sponge',
    'scleromorph sponge': 'Sponge', 
    'lithotelestid coral': 'Stony',
    'gold coral': 'Soft',
    'calcareous sponge': 'Sponge',
    'other coral-like hydrozoan': 'Other'
}

df['CoralType'] = df['VernacularNameCategory'].apply(lambda x: coral_classify[str(x)])

df['Groups Labels']=np.vectorize(lambda x: f'Group {x+1}' if x!=-1 else '')(pad_labels)

def date_period(date_str):
    year = int(date_str[:4]) if type(date_str)==str else 0
    year_mod = year - (year - 20)%50
    return f'{year_mod}-{year_mod+50}' if year_mod>1000 else 'NaN'

df['TimePeriod'] = df['ObservationDate'].apply(date_period)
```

```python
df.to_csv('./processed_deep_sea_corals.gz', compression='gzip')
```
